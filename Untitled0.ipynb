```python
import gradio as gr

def create_website(idea):

    html = f"""
<!DOCTYPE html>
<html>
<head>
<title>{idea}</title>

<style>

body {{
    font-family: Arial;
    margin:0;
    padding:0;
}}

header {{
    background:#4CAF50;
    color:white;
    padding:20px;
    text-align:center;
}}

section {{
    padding:40px;
    text-align:center;
}}

button {{
    padding:10px 20px;
    background:#4CAF50;
    color:white;
    border:none;
    border-radius:5px;
}}

footer {{
    background:#222;
    color:white;
    text-align:center;
    padding:15px;
}}

</style>
</head>

<body>

<header>
<h1>{idea}</h1>
<p>Welcome to our website</p>
</header>

<section>
<h2>About Us</h2>
<p>This website is generated using AI HTML Generator.</p>

<button>Contact Us</button>
</section>

<footer>
<p>© 2026 AI Web Generator</p>
</footer>

</body>
</html>
"""

    return html


app = gr.Interface(
    fn=create_website,
    inputs="text",
    outputs="code",
    title="AI Website Generator",
    description="Enter your website idea and generate HTML page"
)

app.launch()
```
